```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data_df = pd.DataFrame(purchase_data)

purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
player_count = len(purchase_data_df["SN"].unique())
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#find unique items#
items_unique = len(purchase_data_df["Item ID"].unique())
items_unique
#average price of item#
avg_price_item = purchase_data_df["Price"].mean()

#find total revenue#
total_revenue = purchase_data_df["Price"].sum()
total_revenue
#find total number of purchases#
total_purchase = purchase_data_df["Purchase ID"].count()
total_purchase
#make summary table called Purchasing Analysis#

purchase_analysis_df = pd.DataFrame({"Total Unique Items": [items_unique],
                                      "Average Price": [avg_price_item],
                                      "Total Purchases": [total_purchase],
                                      "Total Revenue": [total_revenue]})


purchase_analysis_df["Average Price"] = purchase_analysis_df["Average Price"].map("${:,.2f}".format)
purchase_analysis_df["Total Revenue"] = purchase_analysis_df["Total Revenue"].map("${:,.2f}".format)

purchase_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#find unique values for each gender#

sn_gender_group = purchase_data_df.groupby(["Gender", "SN"])
sn_gender_group.head()

#How many of each gender#
count_per_gender = sn_gender_group["Gender"].value_counts()
count_per_gender


female_total = len(count_per_gender.loc["Female"])
male_total = len(count_per_gender.loc["Male"])
other_total = len(count_per_gender.loc["Other / Non-Disclosed"])


#Percent of each gender#

female_percent = female_total/player_count * 100
male_percent = male_total/player_count * 100
other_percent = other_total/player_count * 100


#Summary table of gender count#
        
gender_data = {"Percentage": [female_percent, male_percent, other_percent], "Total Count": [female_total, male_total, other_total]}

gender_df = pd.DataFrame(data=gender_data, index=["Female", "Male", "Other / Non-Disclosed"])

gender_df["Percentage"] = gender_df["Percentage"].map("{:,.2f}%".format)

gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#find purchase count per gender

gender_purchase = purchase_data_df.groupby(["Gender"])
gender_purchase

purchase_count = gender_purchase["Purchase ID"].count()
purchase_count
#find the average purchase price per gender
avg_price = gender_purchase["Price"].mean()
avg_price

gender_total = gender_purchase["SN"].nunique()
gender_total

#find Total purchase value

total_value = gender_purchase["Price"].sum()
total_value

#average purchase total
avg_purchase_t = (total_value/gender_total).map("${:,.2f}".format)
avg_purchase_t

gen_purchase_df = pd.DataFrame({"Purchase Count": purchase_count,
                                "Avg Purchase Price": avg_price.map("${:,.2f}".format),
                                "Avg Purchase Total": avg_purchase_t})
gen_purchase_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data_df["Age Ranges"] = pd.cut(purchase_data_df["Age"], bins, labels=group_names)

age_range_group= purchase_data_df.groupby("Age Ranges")

total_players_age = age_range_group["SN"].nunique()
total_players_age

pct_players_age = (total_players_age/player_count * 100).map("{:,.2f}%".format)

pct_players_age

age_demo_df = pd.DataFrame({"Percentage": pct_players_age,
                           "Total Players": total_players_age})

age_demo_df


```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Purchase Count
age_purchase_ct = age_range_group["Purchase ID"].count()
age_purchase_ct
#Average Purchase Price
age_avg_price = (age_range_group["Price"].mean()).map("${:,.2f}".format)
age_avg_price
#Total Purchase Value
total_purchase_val = (age_range_group["Price"].sum()).map("${:,.2f}".format)
total_purchase_val
#Average Total Purchase per Person
age_avg_per_player = (age_range_group["Price"].sum()/total_players_age).map("${:,.2f}".format)

age_avg_per_player

purchase_age_df = pd.DataFrame({"Purchase Count": age_purchase_ct,
                   "Average Purchase Price": age_avg_price,
                   "Purchase Totals Per Range": total_purchase_val,
                   "Average Perchase Total Per Person": age_avg_per_player})

purchase_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
sn_group = purchase_data_df.groupby(["SN"])

tot_purchase = sn_group["Purchase ID"].count()
tot_purchase

avg_purch_price = (sn_group["Price"].mean()).map("${:,.2f}".format)
avg_purch_price

totalpurchase_val = (sn_group["Price"].sum()).map("${:,.2f}".format)
totalpurchase_val

top_spenders_df = pd.DataFrame({"Number of Purchases": tot_purchase,
                               "Average Purchase": avg_purch_price,
                               "Total Purchases": totalpurchase_val})
five_top_df = top_spenders_df.sort_values("Total Purchases", ascending=False)
five_top_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Retrieve the Item ID, Item Name, and Item Price columns (using .loc)  
pop_df = purchase_data_df[["Item ID","Item Name","Price"]]
pop_df
# Group by Item ID and Item Name. 
popular_group = pop_df.groupby(["Item ID", "Item Name"])
popular_group

#Purchase count
item_count = popular_group["Item ID"].count()
item_count
#avg item price
item_avg_price = (popular_group['Price'].mean()).map("${:,.2f}".format)
item_avg_price

#total purchse value for item
item_pvalue = popular_group["Price"].sum()
item_pvalue

pop_items_df = pd.DataFrame({"Purchase Count": item_count,
                             "Avg Item Price": item_avg_price,
                             "Total Purchase Value": item_pvalue})
most_pop_items_df = pop_items_df.sort_values("Purchase Count", ascending=False)
most_pop_items_df["Total Purchase Value"] = most_pop_items_df["Total Purchase Value"].map("${:,.2f}".format)
most_pop_items_df.head()
```
